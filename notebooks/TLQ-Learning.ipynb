```python
import numpy as np
from scipy.special import erfc
from scipy.stats import norm
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# --- funciones ---
def f_decision(x,w):	
	fmax_R = x[0]/Rmax
	fmin_ber=np.log(x[1])/np.log(Bermin)
	#fmin_ber=Bermin/x[1]
	#fmin_ber=np.log(Bermin/x[1])
	#fmin_ber=np.log(x[1]/Bermin)
	#fmin_P = Pmin/x[2]
	x=np.array([fmax_R,fmin_ber])
	return np.dot(x,w)
#------------------
```

```python
er=6370 # Radio de la tierra en Km
h=800   # Altura de la órbita en Km
r=er+h    # Distancia del satélite al centro de la tierra 

po=100    #Período orbital en minutos
timestep=9.2 # paso de actualización en segundos
GS=np.array([0,er])# Coordenadas de la "Ground Station"


w0=0.33  # peso del bitrate
w1=1-w0  # peso del ber

at=4 # n° de acciones
alpha =.5
ep=0.1
gamma=0.9
```

```python
k=16973 # cte : PIRE + G/T ver link budget
dmax=3000 # [km]
dmin=800  # [km]
steps=115    # steps

angmax=np.arccos(er/r) # Ángulo en el horizonte
xmax=r*np.sin(angmax)  # Coordenada x en el horizonte ,y=0
dt=(2*np.pi/(po*60))*timestep # Delta ángulo [Rad]

angle=np.arange(-angmax,angmax+dt,dt)
x=r*np.sin(angle) # Coordenada x cada timestep
y=r*np.cos(angle) # Coordenada y cada timestep

d=np.sqrt(x**2+(y-er)**2) # distancia satélite-GS
```

```python
st=len(d)

mt=np.zeros((st,at))
a=mt.T
# matriz de transición del estado i al j según la acción

col=np.arange(1,st+1)
a[:]=col
mt=a.T
```

```python
steps= len(d)
m=4
arg=(k/d)*np.sin(np.pi/m)
ber0=norm.sf(arg)

m=8
arg=(k/d)*np.sin(np.pi/m)
ber1=norm.sf(arg)

m=16
arg=(k/d)*np.sin(np.pi/m)
ber2=norm.sf(arg)

m=32
arg=(k/d)*np.sin(np.pi/m)
ber3=norm.sf(arg)
```

```python
# ber en una arreglo
ba=np.vstack((ber0,ber1,ber2,ber3))
Bermin=min(ber0)
Rmax=4   # accion = m ,1Msb/s para m=4 => R=2 Mb/s
R=np.array([1,2,3,4]) # bit rate ver ITU Bandwith 
```

```python
w0=.33
w1=1-w0
w=np.array([w0,w1])

w0=0.2
w1=1-w0
w=np.array([w0,w1])

rv=np.zeros((steps,at)) # rewards
for n in range(steps):
    for a in range(at):
        baux=ba[a][n]
        #print(ba[a][n],a,n)
        if baux <= Bermin:
            baux=Bermin
        rv[n][a]=1/f_decision([R[a],baux],w)
        #print(R[a]/Rmax/2 + Bermin/baux/2)
pol=np.amax(rv,axis=1)
```

```python
# Linear scalarisation function
def f(v,w):
    return np.dot(v,w)
    
```

```python
# Epsilon greedy action selection from q(s,a)
def e_greedy(qsa):
    z=np.random.uniform(0,1,1)
    if z > ep:
        res = np.where(qsa == max(qsa))
        a=np.random.choice(res[0]) # acción al azar si hay acciones con la misma Qmax(s,a) 
        ac2='Qmax'
    else:
        res = np.where(qsa != max(qsa))
        if res[0].size == 0:
            a=np.random.choice(range(len(qsa))) # en el caso de que todos lo Q(s,a) sean iguales a Qmax(s,a)
        else:
            a=np.random.choice(res[0]) # acción al azar entre las Q(s,a) < Qmax(s,a)
        ac2='Expl'
    return a
    
    
```

```python
# Epsilon greedy action selection from a list of actions
def e_greedy_1(a,aList):
    z=np.random.uniform(0,1,1)
    if z > ep:
        selected_action=a
    else:
        
        selected_action =np.random.choice(aList[aList!=a]) 
    return selected_action
```

```python
def ScalarisedActionSelection(s,Q1,Q2,w):
    ScalarisedQList=np.array([])
    for a in range(at):
        v= [Q1[s, a], Q2[s, a]]
      #  print(v)
        ScalarisedQ=f(v,w)
        #print(ScalarisedQ)
        ScalarisedQList=np.append(ScalarisedQList,ScalarisedQ)
        #print(scalarisedQList)
    return e_greedy(ScalarisedQList) 
```

```python
def mse(x,y):
    return ((x-y)**2).mean()
```

```python
def ThresholdLexicographicActionSelection(s,CQ1,CQ2):
    for a in range(at):
        isSuperior=True
        for aprim in range(at):
            if a != aprim :
                isSuperior=Superior([CQ1[s,a],CQ1[s,aprim],CQ2[s,a],CQ2[s,aprim]] ,1);
            if isSuperior !=True :
                break
        if isSuperior :
            return  e_greedy_1(a,np.arange(at))

```

```python
def Superior(CQ,i):
    if CQ[2*(i-1)] > CQ[2*i-1]:
        return True 
    elif  CQ[2*(i-1)] == CQ[2*i-1]:
        if i==numberOfObjectives:
            return True
        else:
            return Superior(CQ,i+1) 
    else:
        return False
        
    # sigo mañana
```

```python
gamma=0.9
numberOfObjectives=2
C1=3
Q = np.zeros((st+1,at))# el más uno es por que le estoy sumando un estado final 
Q1= np.zeros((st+1,at))
Q2= np.zeros((st+1,at))
CQ1= np.zeros((st+1,at))
CQ2= np.zeros((st+1,at))

np.random.seed(1)
Gacc=np.array([])
racc=np.array([])
N=1000
for e in range(N):
	G=0
	m=0
	ep=0.1
   
	j=0 # j= estado inicial 
	for n in range(steps):
		i=j
		#uj=ScalarisedActionSelection(j,Q1,Q2,w) # elijo la acción e-greedy weigthed average
		uj=ThresholdLexicographicActionSelection(j,CQ1,CQ2); # elijo la acción e-greedy TLQ-Learning
		ui=uj

		j= int(mt[i,ui])     # j next state ,j must be integer if it is used as index
		j=int(j)        # 

		# paso 2	

		Q[i,ui]=Q[i,ui]+alpha*(rv[n][ui] + gamma*max(Q[j])-Q[i,ui]) 
		Q1[i,ui]=Q1[i,ui]+alpha*(r1[n][ui] + gamma*max(Q1[j])-Q1[i,ui]) 
		CQ1[i,ui]=np.min([Q1[i,ui],C1])
		Q2[i,ui]=Q2[i,ui]+alpha*(r2[n][ui] + gamma*max(Q2[j])-Q2[i,ui]) 
		CQ2[i,ui]=Q2[i,ui]
		G+=rv[n][ui]# suma de los rewards
		racc=np.append(racc, rv[n][ui])
		m+=1;
		if m==50000:
			print("break")
			break
	Gacc=np.append(Gacc, G)
#	print("Retornos por episodios de",steps,"pasos Retorno=","{0:0.2f}".format(G),"{0:0.2f}".format(100*G/np.sum(pol)),'%')

print()	
print("máximo retorno","{0:0.2f}".format(np.sum(pol)))
print('ep=',ep)
print('w=',w)

```
