```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create and fit the model
model = LinearRegression()
```

```python
mean=0
std=2
points=100
low=40
high=80
m=3
b=7
x=np.linspace(low,high,points)
y=m*x+b+np.random.normal(mean,std,points)
plt.scatter(x,y)
```

```python
# Create and fit the model
model = LinearRegression()
x=x.reshape(points,1)
y=y.reshape(points,1)
model.fit(X=x, y=y)
plt.scatter(x,y)
plt.plot(x, model.predict(x), color='r')
```

```python
round(model.intercept_[0],2)
```

```python
round(model.coef_[0,0],2)
```

```python
np.sum((y - model.predict(x)) ** 2)
```

```python
 round(model.score(x,y),3)
```
